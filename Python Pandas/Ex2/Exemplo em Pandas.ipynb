```python
import pandas as pd

ca_df = pd.read_excel('ca-2021-02.xlsx')
```

```python
#Exibição das 5 primeiras e 5 últimas linhas
display(ca_df)
```

```python
#Padrão de exibição das 5 primeiras linhas
display(ca_df.head())
```

```python
#Exibir determinada quantidade linhas em sequência
display(ca_df.head(15))
```

```python
#Comando Dataframe.shape
print(ca_df.shape)
#Mostra o número de linhas e colunas (linhas(rows), colunas(columns))

#Apenas linhas
print(ca_df.shape[0])
#Apenas colunas
print(ca_df.shape[1])
```

```python
#Dataframe.info mostra quais são as colunas e o tipo de dados delas
print(ca_df.info())
```

```python
#Dataframe.describe() mostra estatisticas "mais básicas" necessárias
print(ca_df.describe())
```

```python
#Filtrar por uma coluna (utilizando as informações do dataframe.info)
display(ca_df['Revenda'])
```

```python
#Cria-se um novo dataframe com as colunas desejadas
cb_df = ca_df[['Revenda','Municipio','Produto','Valor de Venda']]
display(cb_df)
```

```python
#Exibe as informação da linha desejada (duplo -[[]]- para mais de uma linha)
display(cb_df.loc[[0]])
```

```python
#Exibir uma sequência ordenada especifica
display(cb_df.loc[6:16])
```

```python
#Exibe uma sequência do ponto de partida ao final com o intervalo desejado
display(cb_df.loc[1:11:2])
```

```python
#Crie uma nova dataframe, com as colunas desejadas (dataframe original ou o já modifcado) 
#Escolha o produto (exemplo: gasolina)
gas_df = cb_df.loc[cb_df['Produto'] == 'GASOLINA']
display(gas_df)
```

```python
#Descobrindo qual o maior valor do produto escolhido anteiormente (.max()) ou o menor (.min())
display(gas_df['Valor de Venda'].max())
```

```python
#Exibindo o posto com maior valor, o vendedor e o municipio
display(gas_df[['Revenda','Municipio','Valor de Venda']].max())
```

```python
#DataFrame.loc[] com multiplas condições como filtro (pode-se criar outra dataframe se preferir)
#Quais são os preços nos postos que vendem ETANOL em determinado municipio
#Ordenado pelo menor valor ao maior (inplace=True serve para "salvar" a informação do sort_values)
#Caso use diplay diretamente ao usar o sorrt_values, não é necessário o inplace=True
etosas_df = cb_df.loc[(cb_df['Produto'] == 'ETANOL') & (cb_df['Municipio'] == 'OSASCO')]
etosas_df.sort_values(by='Valor de Venda', inplace=True)
display(etosas_df)
```

```python
#Caso acima sem o inplace
etosas_df = cb_df.loc[(cb_df['Produto'] == 'ETANOL') & (cb_df['Municipio'] == 'OSASCO')]
display(etosas_df.sort_values(by='Valor de Venda'))
```

```python
#Qual a média de preços dos combustíveis GASOLINA e ETANOL no bairro MOOCA em São Paulo
display(ca_df.loc[(ca_df['Bairro'] == 'MOOCA') & 
                    (ca_df['Municipio'] == 'SAO PAULO') & 
                    (ca_df['Produto'].isin(['GASOLINA','ETANOL'])), ['Valor de Venda']].mean())
```

```python
#Outra forma de se fazer o item acima
display(ca_df.loc[(ca_df['Bairro'] == 'MOOCA') & 
                    (ca_df['Municipio'] == 'SAO PAULO') & 
                    ((ca_df['Produto'] == 'GASOLINA') | (ca_df['Produto'] == 'ETANOL')),['Valor de Venda']].mean())
```

```python
#Mostrar a média de valor por combustível no Brasil
display(cb_df[['Produto','Valor de Venda']].groupby(by='Produto').mean())
```

```python
#Acionando um coluna de valor booleano chamada "Ativo" no ca_df que será True para todas as linhas
ca_df['Ativo'] = True
print(ca_df.info())
display(ca_df.head(3))
```
